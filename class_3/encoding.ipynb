# One Hot Encoding 

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np
```

```python
# Dummy categorical data
categories = np.array([['red'], ['green'], ['blue'], ['green'], ['red'], ['blue']])

encoder = OneHotEncoder(sparse_output=False)
encoded = encoder.fit_transform(categories)

print("Original categories:")
print(categories.flatten())
print()
print("One-hot encoded matrix:")
print(encoded)
print()
print("Encoded feature names:")
print(encoder.get_feature_names_out())
```

```python
documents = [
    "people watch movie",
    "people watch cricket",
    "people like movie",
    "people like cricket"
]
```

```python
documents 
```

```python
tokens =[sentence.lower().split() for sentence in documents]
```

```python
tokens
```

```python
all_words= [[words] for sentence in tokens for words in sentence]
```

```python
all_words      # we need to process this data to get the unique words and then we can use one hot encoding to encode the data.
```

```python
encoder =OneHotEncoder(sparse_output=False)
```

```python
encoder.fit(all_words)
```

```python
print("Vocabulary:", encoder.categories_[0])
```

```python
encoder.categories_
```

```python
for sentence in tokens:
    encoded_sentence = encoder.transform([[word] for word in sentence])
    print("\nsentence:", sentence)
    print(encoded_sentence)
```

# Bag of words

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python

documents = ["people watch movie and watch movie again", 
             "people watch cricket and watch cricket",
             "people like movie and like movie a lot",
             "people like cricket"]
```

```python
bow_vector = CountVectorizer()
```

```python
bow_matrix = bow_vector.fit_transform(documents)
```

```python
bow_matrix.toarray()  # this will give us the bag of words representation of the documents. Each row represents a document and each column represents a unique word in the vocabulary. The value in each cell represents the count of the word in the document.
```

```python
print(bow_vector.get_feature_names_out())
print("bow matrix \n", bow_matrix.toarray())
```

```python
print(bow_vector.get_feature_names_out())
```
